```python
# Q3. Compare and contrast mutable and immutable objects in Python with examples.


#In Python, objects can be classified into two categories based on their mutability: mutable 
#and immutable. This distinction affects how objects are modified and how they behave when 
#passed around in a program.

#Mutable Objects
#Definition: Mutable objects are those whose state or value can be changed after they are 
#created. Changes to mutable objects will be reflected in all references to that object.


#Examples of Mutable Objects:

# 1. Lists:

my_list = [1, 2, 3]
my_list.append(4)
print(my_list)  # Output: [1, 2, 3, 4]

#Lists can be modified by adding, removing, or changing elements.


# 2. Sets:

my_set = {1, 2, 3}
my_set.add(4)
print(my_set)  # Output: {1, 2, 3, 4}

#Sets can have elements added or removed.


# 3. Dictionaries:

my_dict = {'a': 1, 'b': 2}
my_dict['c'] = 3
print(my_dict)  # Output: {'a': 1, 'b': 2, 'c': 3}

#Dictionaries can be updated with new key-value pairs or changes to existing ones.


#Behavior with Mutability: If you pass a mutable object to a function and modify it, the 
#changes will affect the original object.



#Immutable Objects
#Definition: Immutable objects are those whose state or value cannot be changed once they 
#are created. Any attempt to modify an immutable object results in the creation of a new 
#object.

#Examples of Immutable Objects:

# 1. Integers

x = 5
x += 1
print(x)  # Output: 6

#Changing the value of x results in the creation of a new integer object.



# 2. Strings:

s = "hello"
s = s.upper()
print(s)  # Output: "HELLO"

#Strings cannot be altered in place. Operations like .upper() return a new string.


# 3. Tuples:

t = (1, 2, 3)
t = t + (4,)
print(t)  # Output: (1, 2, 3, 4)

#Tuples cannot be modified once created. Any modification results in the creation of a new tuple.


#Behavior with Immutability:


#If you pass an immutable object to a function and "modify" it, you actually create and 
#return a new object.

def modify_string(s):
    s = s.upper()
    return s

my_str = "hello"
my_str = modify_string(my_str)
print(my_str)  # Output: "HELLO"



#Key Differences

# 1. Mutability:

#Mutable: Can be changed in place. Examples: Lists, Dictionaries, Sets.
#Immutable: Cannot be changed in place. Examples: Integers, Strings, Tuples.

# 2. Behavior in Functions:

#Mutable: Changes made in functions affect the original object.
#Immutable: Changes create a new object, leaving the original unchanged.

# 3. Performance:

#Mutable: Generally more memory-efficient for modifications because changes are made in place.
#Immutable: May have more overhead because new objects are created for each modification.


#Understanding mutable and immutable objects is crucial for managing state and behavior in 
#Python programs, as it affects how data is handled and shared across different parts of the 
#code.





```

```python
# Q4. Discuss the different types of operators in Python and provide examples of how they are 
#used?



#In Python, operators are special symbols or keywords that perform operations on values or 
#variables. Python supports several types of operators, each serving different purposes. 
#Here’s a rundown of the main types of operators in Python and examples of how they are used:

# 1. Arithmetic Operators:
#These operators are used for mathematical operations.

# Addition (+):

a = 5
b = 3
result = a + b
print(result)  # Output: 8


# Subtraction (-):

a = 5
b = 3
result = a - b
print(result)  # Output: 2


# Multiplication (*):

a = 5
b = 3
result = a * b
print(result)  # Output: 15


# Division (/):

a = 5
b = 2
result = a / b
print(result)  # Output: 2.5


# Integer Division (//):

a = 5
b = 2
result = a // b
print(result)  # Output: 2


# Modulus (%):

a = 5
b = 2
result = a % b
print(result)  # Output: 1


# Exponentiation (**):

a = 2
b = 3
result = a ** b
print(result)  # Output: 8



#2. Comparison Operators

#These operators are used to compare two values.

# Equal to (==):

a = 5
b = 5
result = (a == b)
print(result)  # Output: True


# Not equal to (!=):

a = 5
b = 3
result = (a != b)
print(result)  # Output: True


# Greater than (>):

a = 5
b = 3
result = (a > b)
print(result)  # Output: True


# Less than (<):

a = 5
b = 3
result = (a < b)
print(result)  # Output: False


# Greater than or equal to (>=):

a = 5
b = 5
result = (a >= b)
print(result)  # Output: True


# Less than or equal to (<=):

a = 5
b = 6
result = (a <= b)
print(result)  # Output: True


# 3. Logical Operators

#These operators are used to combine conditional statements.

# AND (and):

a = True
b = False
result = (a and b)
print(result)  # Output: False

# OR (or):

a = True
b = False
result = (a or b)
print(result)  # Output: True


# NOT (not):

a = True
result = not a
print(result)  # Output: False


# 4. Assignment Operators

#These operators are used to assign values to variables.

# Assign (=):

a = 5


#Add and assign (+=):

a = 5
a += 3
print(a)  # Output: 8



#Subtract and assign (-=):


a = 5
a -= 2
print(a)  # Output: 3


# Multiply and assign (*=):


a = 5
a *= 2
print(a)  # Output: 10


# Divide and assign (/=):


a = 5
a /= 2
print(a)  # Output: 2.5


# Modulus and assign (%=):


a = 5
a %= 2
print(a)  # Output: 1


# Exponent and assign (**=):


a = 2
a **= 3
print(a)  # Output: 8



# 5. Bitwise Operators

#These operators perform bit-level operations.

# AND (&):


a = 5  # 0101 in binary
b = 3  # 0011 in binary
result = a & b
print(result)  # Output: 1 (0001 in binary)


# OR (|):


a = 5  # 0101 in binary
b = 3  # 0011 in binary
result = a | b
print(result)  # Output: 7 (0111 in binary)


# XOR (^):


a = 5  # 0101 in binary
b = 3  # 0011 in binary
result = a ^ b
print(result)  # Output: 6 (0110 in binary)


# NOT (~):


a = 5  # 0101 in binary
result = ~a
print(result)  # Output: -6 (bitwise NOT of 0101)


# Left Shift (<<):


a = 5  # 0101 in binary
result = a << 1
print(result)  # Output: 10 (1010 in binary)


# Right Shift (>>):


a = 5  # 0101 in binary
result = a >> 1
print(result)  # Output: 2 (0010 in binary)


# 6. Membership Operators

#These operators check for membership in sequences.


# In (in):


my_list = [1, 2, 3]
result = 2 in my_list
print(result)  # Output: True


# Not in (not in):


my_list = [1, 2, 3]
result = 4 not in my_list
print(result)  # Output: True


# 7. Identity Operators

#These operators are used to compare the memory locations of two objects.


# Is (is):


a = [1, 2, 3]
b = a
result = (a is b)
print(result)  # Output: True


# Is not (is not):


a = [1, 2, 3]
b = [1, 2, 3]
result = (a is not b)
print(result)  # Output: True



# Understanding these operators and their usage helps in performing a wide range of operations in Python, 
#from simple arithmetic to complex logic and data manipulation.

```

```python
# Q7. Describe the different types of loops in Python and their use cases with examples.





# In Python, loops are essential for iterating over a sequence of items or repeating a block of code multiple 
# times. Python primarily provides two types of loops: for loops and while loops. Here’s a breakdown of each 
# type along with their use cases and examples:

# 1. for Loop

# The for loop is used to iterate over a sequence (like a list, tuple, string, or range) and execute a block 
# of code for each item in the sequence. It's particularly useful when the number of iterations is known or 
# when iterating over items in a collection.

# Basic Syntax:

for variable in sequence:
    # Code to execute

    
# Use Cases:

# Iterating through elements of a list, tuple, or other iterable.

# Performing an action a specific number of times using range().

# Iterating over characters in a string.


# Examples:

# 1. Iterating through a List:

fruits = ['apple', 'banana', 'cherry']
for fruit in fruits:
    print(fruit)

    
# Output:

apple
banana
cherry


# 2. Using range() to Iterate a Fixed Number of Times:


for i in range(5):
    print(i)

    
    
# Output:


0
1
2
3
4


# 3. Iterating through a String:


for char in "hello":
    print(char)

    
# Output:

h
e
l
l
o



# 4. Iterating with enumerate():


colors = ['red', 'green', 'blue']
for index, color in enumerate(colors):
    print(index, color)

    
# Output:

0 red
1 green
2 blue






# 2. while Loop

# The while loop repeatedly executes a block of code as long as a given condition remains True. It’s useful 
# when the number of iterations is not known beforehand and depends on dynamic conditions.

# Basic Syntax:

while condition:
    # Code to execute

    
# Use Cases:
# When you want to execute code as long as a condition remains true.
# When the number of iterations cannot be determined in advance.


# Examples:

# 1. Basic while Loop:
 
count = 0
while count < 5:
    print(count)
    count += 1

    
# Output:

0
1
2
3
4



# 2. Loop with a Break Condition:

number = 0
while True:
    print(number)
    number += 1
    if number >= 3:
        break

        
# Output:

0
1
2



# 3. Using while with User Input:


user_input = ''
while user_input.lower() != 'exit':
    user_input = input("Type 'exit' to stop: ")
    print(f"You typed: {user_input}")

    
# Output:

Type 'exit' to stop: hello
You typed: hello
Type 'exit' to stop: exit
You typed: exit


# Summary

# for Loop: Ideal for iterating over sequences and when the number of iterations is known or fixed. Commonly 
# used with collections and ranges.

# while Loop: Suited for cases where the number of iterations depends on dynamic conditions or when the loop 
# should continue until a condition changes.


# Both types of loops are versatile and can be used in various scenarios to control the flow of execution in 
# your programs.


```
